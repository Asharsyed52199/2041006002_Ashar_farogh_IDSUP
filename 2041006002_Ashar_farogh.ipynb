```python
import random
import math
import numpy as np
from collections import Counter 
import matplotlib.pyplot as plt
```

```python
age = []
for i in range(1000):
    r=random.randrange(60,70)
    age.insert(i,r)
age.sort()
print(age)
```

```python
age_mean = np.average(age)#E(x)
print('\nPopulation mean is',age_mean)
age_std = np.std(age)#sigma
print('Population standard deviation is', age_std,'\n')
```

```python
Counter = Counter(age)
X , Y = [] , []
for i in Counter.keys() :
    X.append(i)
for j in Counter.values():
    Y.append(j)
#print(X)
#print(Y)

plt.hist(age ,edgecolor='black',color='grey')
plt.show()
```

```python

sample = []
sample_mean = []

for i in range(500):
      for j in range(30):
        sample.append(random.choice(age))
        #print('\n',sample)
      sm = np.average(sample)
      sample_mean.append(sm)
print(sample_mean)
```

```python
plt.hist(sample_mean ,edgecolor='blue',color='cyan')
```

```python
M = np.average(sample_mean)
print("\nMean of the sample_mean:",M)
print("Is the mean of s_mean[] same as age_mean or E(x) ?")
print( M is age_mean,"\n")

N = np.std(sample_mean)/math.sqrt(30)
print("\nMean of the sample_mean:",N)
print("Is the standard deviation of sample_mean[] = age_std/root30(n=30) ?")
print(N is age_std )
```
